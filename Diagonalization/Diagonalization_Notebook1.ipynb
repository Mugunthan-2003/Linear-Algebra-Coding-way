Basis vectors are independent vectors and all other vectors in a plane can expressed as a linear combination of basis vectors .

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Standard basis vectors in R^2
e1 = np.array([[1], [0]]) 
e2 = np.array([[0], [1]])

# diagonal matrix 
D = np.array([[3, 0], [0, 6]]) 
```

```python
def transform(x):
    """Apply the transformation D to the vector x.

    Parameters:
        x(vector): A 2D vector

    Returns:
        Points of the vector x and the transformed vector
    """
    D = np.array([[3, 0], [0, 6]])
    x_xaxis, x_yaxis = x.ravel()
    result = D @ x # Apply the transformation
    result_xaxis, result_yaxis = result.ravel()
    return x_xaxis, x_yaxis, result_xaxis, result_yaxis
```

```python
e1_x, e1_y, trans_e1_x, trans_e1_y = transform(e1)
e2_x,e2_y,trans_e2_x, trans_e2_y = transform(e2)
# Create the plot
plt.figure(figsize=(6, 6))  # Adjust figure size as needed

# Plot arrows with custom head width and length
plt.arrow(0, 0, e1_x, e1_y, head_width=0.1, head_length=0.1, color='green')
plt.arrow(0, 0, trans_e1_x, trans_e1_y, head_width=0.1, head_length=0.1, color='blue')
plt.arrow(0, 0, e2_x, e2_y, head_width=0.1, head_length=0.1, color='red')
plt.arrow(0, 0,trans_e2_x, trans_e2_y, head_width=0.1, head_length=0.1, color='yellow')

# Add text labels slightly above the arrow tips (adjust offsets as needed)
plt.text(e1_x + 0.1, e1_y + 0.1, 'e1', color='red', fontsize=12)
plt.text(trans_e1_x + 0.1, trans_e1_y + 0.1, 'T(e1)', color='blue', fontsize=12)
plt.text(e2_x + 0.1, e2_y + 0.1, 'e2', color='green', fontsize=12)
plt.text(trans_e2_x + 0.1, trans_e2_y + 0.1, 'T(e2)', color='yellow', fontsize=12)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of e1 and e2 vectors and their transformation T(e1) and T(e2)')

# Display the plot
plt.grid(True)
plt.show()
```

On applying Transformation in standard basis vectors , It is restricted to reach to vectors only in that axis and not to other parts .

```python
vector = np.array([[1], [1]])
vector_x, vector_y, trans_vector_x, trans_vector_y = transform(vector)

# Create the plot
plt.figure(figsize=(6, 6))  # Adjust figure size as needed

# Plot arrows with custom head width and length
plt.arrow(0, 0, vector_x, vector_y, head_width=0.1, head_length=0.1, color='green')
plt.arrow(0, 0, trans_vector_x, trans_vector_y, head_width=0.1, head_length=0.1, color='blue')


# Add text labels slightly above the arrow tips (adjust offsets as needed)
plt.text(vector_x + 0.1, vector_y + 0.1, 'vector', color='red', fontsize=12)
plt.text(trans_vector_x + 0.1, trans_vector_y + 0.1, 'T(vector)', color='blue', fontsize=12)


# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of vector({},{}) and T(vector)'.format(vector_x,vector_y))

# Display the plot
plt.grid(True)  # Add grid lines (optional)
plt.show()
```

On applying Transformation in other vectors , all parts of the plane can be reached .

```python
vector = np.array([[5], [7]])
vector_x, vector_y, trans_vector_x, trans_vector_y = transform(vector)
# Create the plot
plt.figure(figsize=(6, 6))  # Adjust figure size as needed

# Plot arrows with custom head width and length
plt.arrow(0, 0, vector_x, vector_y, head_width=0.1, head_length=0.1, color='green')
plt.arrow(0, 0, trans_vector_x, trans_vector_y, head_width=0.1, head_length=0.1, color='blue')


# Add text labels slightly above the arrow tips (adjust offsets as needed)
plt.text(vector_x + 0.1, vector_y + 0.1, 'vector', color='red', fontsize=12)
plt.text(trans_vector_x + 0.1, trans_vector_y + 0.1, 'T(vector)', color='blue', fontsize=12)


# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of vector({},{}) and T(vector)'.format(vector_x,vector_y))

# Display the plot
plt.grid(True)  # Add grid lines (optional)
plt.show()
```

```python
# diagonal matrix 
D = np.array([[3, 0], [0, 6]]) 
square=D@D
cube=D@D@D
print("Diagonal matrix :\n",D)
print("Square of D:\n",square)
print("Cube of D:\n",cube)
```

On multiplying Diagonal matrix n times , it's elements get updated by power of n .
