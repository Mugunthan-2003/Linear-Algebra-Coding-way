Steps :

1. Equations are formulated
2. Equations are converted to matrix
3. Corresponding Eigen values and vectors are found
4. The eigen vector whose eigen value is 1 is found and vector is scaled to sum of 1. (At steady state Ax=A ,implicitly lambda=1) .

Total rank of a page is sum of the ranks of incoming nodes .
Assuming

x1 be rank of page 1

x2 be rank of page 2

x3 be rank of page 3

x4 be rank of page 4.



so the equations are

x1=x3+(x4/2)

x2=(x1/3)

x3=(x1/3)+(x2/2)+(x4/2)

x4=(x1/3)+(x2/2)

```python
import numpy as np # array operations
```

```python
Adj_matrix = [
    [0, 0, 1, 1/2],
    [1/3, 0, 0, 0],
    [1/3, 1/2, 0, 1/2],
    [1/3, 1/2, 0,0]
    ]                    # adjacency matrix
```

```python
eigen_values, eigen_vectors = np.linalg.eig(Adj_matrix) # eigenvalues and eigenvectors of the matrix
```

```python
Index=0 # index of the eigenvector corresponding to eigenvalue
for i in eigen_values:
    if(round(i.real,2)==1):# checking if the eigenvalue is 1
        significant = eigen_vectors[:,Index].real # storing the eigenvector
        print("Eigenvector corresponding to eigenvalue 1 is: ",significant)
        break
    Index+=1 # incrementing the index
```

```python
significant=significant.real/sum(significant.real) # scaling the significant vector
significant=significant.round(3) # rounding off to 3 decimal places
```

```python
print("The page rank of the pages are: ")   # printing the page rank of the pages
for i in range(len(significant)):
    print("x{}(Page{}) ==> {}".format(i+1,i+1,significant[i]))
```
