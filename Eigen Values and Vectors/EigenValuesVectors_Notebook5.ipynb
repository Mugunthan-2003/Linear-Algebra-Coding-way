To obtain the ranks of a page :
1. Initialize the starting deterministic vector
2. Multiply the vector consectively with adjacent matrix
3. stop the above multiplication when no changes are observed in vector .

Now the vector x is called " Steady state vector "

```python
import numpy as np
```

```python
Adj_matrix = np.array([
    [0, 0, 1, 1/2],
    [1/3, 0, 0, 0],
    [1/3, 1/2, 0, 1/2],
    [1/3, 1/2, 0,0]
    ])                    # adjacency matrix
```

```python
def page_rank(Adj_matrix, page,iteration): # function to calculate the page rank
    """This function calculates the page rank of the pages using the adjacency matrix.

    Parameters:
        Adj_matrix(matrix): The adjacency matrix of the pages.
        page(int): The starting page number.
        iteration(int): The number of iterations to be performed.
    """
    x=[1 if i==page-1 else 0 for i in range(len(Adj_matrix))] # initial vector
    print("Starting page: ",page)
    x=np.array(x) # converting the list to array
    print("Initial vector: ",x)
    for i in range(iteration):
        x=np.dot(Adj_matrix,x) # multiplying the adjacency matrix with the vector
    print("The page rank of the pages after {} iterations : ".format(iteration))   # printing the page rank of the pages
    for i in range(len(x)):
        print("Page{} ==> {}".format(i+1,round(x[i],3)))
```

```python
page_rank(Adj_matrix, 1,100) # calling the function with the adjacency matrix and the page number
```

```python
page_rank(Adj_matrix, 3,100) # calling the function with the adjacency matrix and the page number
```
